# ðŸ’³ AnÃ¡lise de CrÃ©dito com AWS S3 + Athena

Este projeto faz parte do curso **ProfissÃ£o Analista de Dados** da EBAC. Utilizamos dados simulados de crÃ©dito, armazenamos no **AWS S3**, criamos a tabela no **AWS Athena**, executamos consultas SQL e por fim analisamos os resultados com Python no Google Colab.

A seguir, apresento as etapas e insights extraÃ­dos a partir dessas anÃ¡lises.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
````

## ðŸ“Œ Consulta 1 â€“ Quantidade de clientes por faixa de renda

**Consulta SQL no Athena:**
```sql
SELECT renda_mensal, COUNT(*) AS total
FROM credito8
GROUP BY renda_mensal
ORDER BY total DESC;
```

VisualizaÃ§Ã£o dos dados:

````python
df_renda = pd.DataFrame({
    "renda_mensal": ["menos que $40K", "$40K - $60K", "$60K - $80K", "$80K - $120K", "mais que $120K"],
    "total": [150, 120, 95, 70, 22]
})

plt.figure(figsize=(10,6))
sns.barplot(x="total", y="renda_mensal", data=df_renda, palette="viridis")
plt.title("DistribuiÃ§Ã£o de Clientes por Faixa de Renda")
plt.xlabel("Quantidade")
plt.ylabel("Faixa de Renda")
plt.tight_layout()
plt.show()
````

## ðŸ“Œ Consulta 2 â€“ MÃ©dia do Score por faixa de renda

**Consulta SQL no Athena:**
```sql
SELECT renda_mensal, AVG(score) AS media_score
FROM credito8
GROUP BY renda_mensal
ORDER BY media_score DESC;
```

VisualizaÃ§Ã£o dos dados:

````python
df_score = pd.DataFrame({
    "renda_mensal": ["menos que $40K", "$40K - $60K", "$60K - $80K", "$80K - $120K", "mais que $120K"],
    "media_score": [520, 570, 610, 655, 700]
})

plt.figure(figsize=(10,6))
sns.lineplot(x="renda_mensal", y="media_score", data=df_score, marker='o')
plt.title("Score MÃ©dio por Faixa de Renda")
plt.xlabel("Faixa de Renda")
plt.ylabel("Score MÃ©dio")
plt.tight_layout()
plt.show()
````

## âœ… ConclusÃ£o

- A maior parte dos clientes estÃ¡ na faixa de renda "menos que $40K"
- Clientes com maior renda tendem a ter maior score de crÃ©dito
- Isso pode indicar menor risco de inadimplÃªncia nas faixas mais altas
